```python
import pandas as pd
import mysql.connector
from sqlalchemy import create_engine
from urllib import request
from urllib.parse import quote_plus
```

```python
mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="root@123",
  database="data_lake"
)
```

```python
engine = create_engine('mysql+mysqlconnector://root:%s@localhost/data_lake' % quote_plus("root@123"))
```

```python
# Read the CSV file into a pandas DataFrame
df = pd.read_csv("data_lake.csv")
```

```python
# Slice the data by selecting a subset of columns and/or rows
subset = df.loc[29292:30584, ['ranks', 'title', 'no_of_reviews', 'ratings', 'author', 'cover_type', 'year','genre', 'price']]
```

```python
subset.shape
```

```python
subset.info()
```

```python
# Check for missing values in each column
subset.isnull().sum()
```

```python
# Drop rows with missing values
subset = subset.dropna()
subset
```

```python
# use reindex to change the order of columns
subset = subset.reindex(columns=["ranks","title","no_of_reviews","ratings","author", "year","price", "genre", "cover_type"])
subset
```

```python
subset = pd.DataFrame(subset)
subset.columns = ["book_rank", "book_title", "no_of_reviews", "rating","authors", "year", "price", "genre", "cover_type"]

subset
```

```python
# Set the index of the DataFrame to a sequence of integers
subset.reset_index(drop=True, inplace=True)

subset
```

```python
# check for null values in the 'price' column
null_values = subset['price'].isnull().sum()

print("The 'price' column has", null_values, "null values.")
```

```python
# Replace NaN values with 0
subset['year'] = subset['year'].fillna(0)

# Convert column to integers
subset['year'] = subset['year'].astype(int)

subset
```

```python
# Replace NaN values with 0
subset['book_rank'] = subset['book_rank'].fillna(0)

# Convert column to integers
subset['book_rank'] = subset['book_rank'].astype(int)

subset
```

```python

#subset.to_sql(name='books', con=mydb, if_exists='replace', index=False)
subset.to_sql("books", con = engine, if_exists= 'replace', index=False )
```
